SUM OF NUMBERS IN THE GIVEN RANGE

```python
# Using logic
def give_sum(lower_limit,upper_limit):
    sum = 0
    for i in range(lower_limit,upper_limit+1):
        sum += i
    return sum
```

```python
num_l = int(input('Enter the lower limit: '))
num_u = int(input('Enter the upper limit: '))

ans = give_sum(num_l,num_u)
print(f'The sum from {num_l} to {num_u} is {ans}')
```

```python
# Using formula
def give_sum_2(num1,num2):
    # Refer to prepinsta for the formula
    return ((num2*(num2+1)/2) - (num1*(num1+1)/2) + num1)
```

```python
give_sum_2(2,5)
```

```python
give_sum_2(1,100)
```

GREATEST OF TWO NUMBERS

```python
def greater(num1,num2):
    if num1 > num2:
        return num1
    return num2
```

```python
a = int(input('Enter the first number: '))
b = int(input('Enter the second number: '))
print(f'Greater among {a} and {b} ----> {greater(a,b)}')
```

```python
# Using in-built functions
a,b,c = input('Enter the three numbers: ').split()
a = int(a) ; b = int(b) ; c = int(c)
l = [a,b,c]
print(f'The greatest among the numbers are {max(l)}')
```

```python
# Using own logic
def greater_among_three(a,b,c):
    if a > b:
        if a > c:
            return a
        return c
    elif b > a:
        if b > c:
            return b
        return c
```

```python
a = int(input('Enter first number: '))
b = int(input('Enter second number: '))
c = int(input('Enter third number: '))

print(f'Max number among these is {greater_among_three(a,b,c)}')
```

! LEAP YEAR !

```python
def check_leap(year):
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        return True
    return False
```

```python
y = int(input('Enter the year to check if leap year: '))
if check_leap(y):
    print(f'{y} is a leap year')
else:
    print(f'{y} is not a leap year')
```

! PRIME NUMBER !

```python
def check_prime(prime):
    for i in range(2,int(prime/2) + 1):
        if prime % i == 0:
            return False
    return True
```

```python
num = int(input('Enter the number to check prime: '))
if check_prime(num):
    print(f'{num} is a prime number!')
else:
    print(f'{num} is not a prime number!')
```

PRINT PRIME NUMBERS WITHIN A GIVEN RANGE

```python
def check_prime(prime):
    for i in range(2,int(prime/2) + 1):
        if prime % i == 0:
            return False
    return True

def give_primes(a,b):
    numbers = []
    for i in range(a,b+1):
        if check_prime(i):
            numbers.append(i)
    return numbers
```

```python
give_primes(1,10)
```

```python
def primes(a,b):
    prime_list = []
    for i in range(a,b+1):
        flag = 0
        if i < 2:
            continue
        if i == 2:
            prime_list.append(i)
            continue
        for j in range(2,i):
            if i % j == 0:
                flag = 1
                break
        if flag == 0:
            prime_list.append(i)
    return prime_list
```

```python
primes(1,50)
```

SUM OF DIGITS OF A NUMBER

```python
def give_sum(num):
    num = str(num)
    number_list = [int(x) for x in num]
    sum_of_nos = 0
    for i in number_list:
        sum_of_nos += i
    return sum_of_nos
```

```python
give_sum(12345678987654321)
```

! REVERSING A NUMBER !

```python
# By using logic
def get_reverse(num):
    rev = 0
    while num != 0:
        rem = num % 10
        rev = rev * 10 + rem
        num = num // 10
    return rev
```

```python
get_reverse(1234)
```
